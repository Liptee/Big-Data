**С использованием *Zookeeper* сервиса решите проблему обедающих философов (каждый философ - отдельный процесс в системе)** 

```python
from time import time, sleep
from kazoo.client import KazooClient
from threading import Thread
import random
```

```python
class ForkState():
    free = 1
    busy = 0

class Fork():
    def __init__(self):
        self._state = ForkState.free
        self._owner = None

    @property
    def owner(self):
        return self.owner

    @owner.setter
    def owner(self, owner):
        self._owner = owner

    @property
    def state(self):
        return self._state

    @state.setter
    def state(self, state):
        self._state = state

class State():
    eating = 1
    thinking = 0

class Philosopher(Thread):
    def __init__(self, name, resource):
        self._name = name
        self._state = State.thinking
        self._eating_time = 5
        self._resource_pointer = [Fork()]
        Thread.__init__(self)

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, name):
        self._name = name

    @property
    def state(self):
        return self._state

    @state.setter
    def state(self, state):
        self._state = state

    def eat(self):
        self.state(State.eating)
        print(f"Философ {self.name} ест")
        sleep(self._eating_time)
        for f in Fork:
            f.state = Fork.free
            f.owner = None

    def think(self):
        self.state = State.thinking
        print(f"Философ {self.name} размышляет")
        sleep(random.randint(1,10))

    def run(self):
        zk = KazooClient()
        zk.start()
        start = time()

        print(f"Философ {self.name} сел за стол")
        while time() - start < N:
            acq_chopsticks = []
            for f in self._resource_pointer:
                if Fork.state == ForkState.free:
                    Fork.state == ForkState.busy
                    Fork.owner = self.name
                    acq_chopsticks.append(Fork)
                    if len(acq_chopsticks) == 2:
                        self.eat()
                else:
                        self.think()
        zk.stop()
        zk.close()
```

```python
master_zk = KazooClient()
master_zk.start()

if master_zk.exists('/task1'):
    master_zk.delete('/task1', recursive=True)

master_zk.create('/task1')
master_zk.create('/task1/table')
master_zk.create('/task1/forks')
master_zk.create('/task1/forks/1')
master_zk.create('/task1/forks/2')
master_zk.create('/task1/forks/3')
master_zk.create('/task1/forks/4')
master_zk.create('/task1/forks/5')

fork_array = 'forks'
N=20

for i in range(5):
    p = Philosopher(i, fork_array)
    p.start()
```
